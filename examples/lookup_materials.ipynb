## How to look for materials in the database?
We can use the function `find_material` to list all materials satisfying certain constraints:

```python
from mammos_dft.db import find_materials
```

```python
find_materials()
```

```python
find_materials(chemical_formula="CrNiP")
```

```python
find_materials(cell_angle_gamma="90 deg")
```

```python
import math

find_materials(cell_angle_gamma=f"{math.pi/2} rad")
```
